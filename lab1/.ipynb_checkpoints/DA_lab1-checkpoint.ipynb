### Import data and  library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Data1.csv',index_col=0)
```

```python
print(df)
```

### Plot of imported data

```python
df.plot()
```

### Conversion DataFrame by using melt function
It allows to change format of DataFram, now first colunm is index, second is type of Theta and the third one is Value.

```python
dfmelt = pd.melt(df,var_name='Theta',value_name='Value')
print(dfmelt)
```

### Histogram plot
Previously created data frame, allows to plot histograms depending on theta type. By using seaborn library and FacetGrid, 6 historam was created, each one for one theta. 

Based on plots, we can observe:
- In case of theta_1, we have consistenet distribution, resembling a normal distribution with expected value 0
- In case of theta_2, we have similar situation as in last case, but with bigger variance
- In case of theta_3, it is same as, in last two cases, but variance is much lower
- In case of theta_4, it is very inconsistent distribution, it's look a litlle bit as normal inverse distribution
- In case of theta_5, it is also inconsistent distribution, with 3 peaks at -1,0,1 values
- In case of theta_6, it is again norlam distribution, but with expected value 5, and variance bigger than in cases of theta 1-3

```python
import seaborn as sns

g = sns.FacetGrid(dfmelt,row="Theta",sharex=False)
g.map(plt.hist,'Value')
```

### Analysis for theta 1-4 in 2018

Only the required data has been selected, by using loc.

```python
df2018 = df['2018-01-01':'2018-12-31']
df2018t = df2018.loc[:,'theta_1':'theta_4']
print(df2018t)
```

We can observe, that data in 2018 are more consistent, then in case of all measured time.

```python
df2018t.plot(kind='kde')
```
